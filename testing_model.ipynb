```python
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from deep_translator import GoogleTranslator
import pickle

# Load model
model_path = '/content/drive/MyDrive/Colab Notebooks/Resha Ananda Rahman (ABSA)/Model/distilbert_model.pkl'
with open(model_path, 'rb') as f:
    model = pickle.load(f)

# Load tokenizer
tokenizer_path = '/content/drive/MyDrive/Colab Notebooks/Resha Ananda Rahman (ABSA)/Model/distilbert_tokenizer.pkl'
with open(tokenizer_path, 'rb') as f:
    tokenizer = pickle.load(f)

# Inisialisasi DeepL Translate
translator = GoogleTranslator(source='id', target='en')

# Input teks baru dalam bahasa Indonesia
new_text_id = input("\nMasukkan teks baru (bahasa Indonesia): ")

# Terjemahkan teks ke bahasa Inggris
new_text_en = translator.translate(new_text_id)

# Tokenisasi teks bahasa Inggris
inputs = tokenizer(new_text_en, return_tensors="pt")

# Prediksi dengan model
with torch.no_grad():
    logits = model(**inputs).logits

predicted_class_id = logits.argmax().item()
predicted_label = model.config.id2label[predicted_class_id]

print(f"\nTeks Bahasa Indonesia: {new_text_id}")
print(f"\nTeks Bahasa Inggris (diterjemahkan): {new_text_en}")
print(f"\nPrediksi Sentimen: {predicted_label}")
```
